```python
# Importing modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Declaring a list
my_list = [1, 2, 3, 4, 5]

# Declaring a NumPy array
my_array = np.array([1, 2, 3, 4, 5])

# Declaring a basic Python function
def square(x):
    return x**2

# Creating a simple Matplotlib figure
x = np.linspace(0, 10, 100)  # Generate 100 evenly spaced points from 0 to 10
y = np.sin(x)               # Compute the sine of each point

```

```python
# Create a plot
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Save the figure as a PNG file
plt.savefig("sine_wave.png")

# Show the plot (optional)
plt.show()
```
